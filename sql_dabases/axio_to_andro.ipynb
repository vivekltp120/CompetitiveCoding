```python
import pandas as pd
axio_df=pd.read_csv("./expenses/axio_expense_report_01-03-2020_to_31-01-2025.csv")
axio_new=axio_df.iloc[4:,].reset_index()
axio_new.columns=axio_new.iloc[0]
axio_new.drop(labels=0,axis=0,inplace=True)
axio_new.drop(columns=[4],inplace=True)
axio_new.head(5)
```

```python
#getting data from the andromoney
andro_df=pd.read_csv("./expenses/AndroMoney_walnut.csv")
#change the dtaetime formate
# andro_df['Date'] = andro_df['Date'].apply(lambda x: pd.to_datetime(str(x), format='%Y%m%d'))
andro_new=andro_df.iloc[17:,].reset_index()
andro_new['Date'] = pd.to_datetime(andro_new['Date'].astype(str), format='%Y%m%d',errors='coerce')
andro_new['Date'].info()
andro_new.head(5)
```

```python
#creating a column on the basis of other column
def drcr(row):
    if isinstance(row["Expense(Transfer Out)"],str):
        return "DR"
    else:
        return "CR"
andro_new["DR/CR"] = andro_new.apply(drcr,axis=1)
andro_new.drop(columns=['index'],inplace=True)
andro_new.head(5)
```

```python
def add_hash(row):
    # print(row)
    return "".join(["#",row['Sub-Category']])

andro_new['Sub-Category']=andro_new.apply(add_hash,axis=1)
andro_new.head(10)

```

```python
# Condition operator
temp=andro_new[(andro_new["Category"]!='Income') & (andro_new["Amount"]>5000)]
print(temp.head)
temp2=andro_new.query("Category!='Income' and Amount>5000")
print(temp2.head)
if all(temp)==all(temp2):
    print("Same")
```
